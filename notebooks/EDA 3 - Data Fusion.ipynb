### Extrair dados

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = "./dados/caso_estudo.xlsx"
```

```python
dfClientes = pd.read_excel(dataset, sheet_name='clientes')
dfLojas = pd.read_excel(dataset, sheet_name='lojas')
dfProdutos = pd.read_excel(dataset, sheet_name='produtos')
dfVendas = pd.read_excel(dataset, sheet_name='vendas')
dfPag = pd.read_excel(dataset, sheet_name='pagamentos')
```

### Primary Key

```python
dfClientes = dfClientes.set_index("id")
dfLojas = dfLojas.set_index("id")
dfProdutos = dfProdutos.set_index("id")
dfVendas = dfVendas.set_index("id")
dfPag = dfPag.set_index("id")
```

### Join

```python
df = dfVendas.join(dfClientes.add_prefix("cliente_"), on="id_cliente")
df = df.join(dfLojas.add_prefix("loja_"), on="id_loja")
df = df.join(dfProdutos.add_prefix("produto_"), on="id_produto")

df
```

```python
df = df.join(dfPag.set_index("id_venda"))
df
```

## Feature Engineering

```python
df["pg"] = 1
df.loc[df.dt_pgto.isnull(), "pg"] = 0
df
```

```python
df["tempo_pg"] = (df.dt_pgto - df.dt_venda).dt.days
df
```

```python
df.cliente_dt_nasc = pd.to_datetime(df.cliente_dt_nasc, format = "%m/%d/%Y")
```

```python
import numpy as np
df["cliente_idade"] = np.floor((pd.to_datetime("today") - df.cliente_dt_nasc).dt.days / 365)
df
```
